```python
# TP Data Mining - Collecte et fusion de données multi-sources

# 1. Importation des bibliothèques nécessaires
import pandas as pd
import requests

# 2. Lecture des données depuis un fichier CSV
etudiants = pd.read_csv("student_performance.csv")
print("Données du fichier CSV :")
print(etudiants.head())

# 3. Lecture des données depuis un fichier Excel
excel_data = pd.read_excel("data.xlsx")
print("\nDonnées du fichier Excel :")
print(excel_data.head())

# 4. Lecture des données depuis un fichier JSON
movies = pd.read_json("movies.json")
print("\nDonnées du fichier JSON :")
print(movies.head())

# 5. Récupération de données depuis une API
url = "https://jsonplaceholder.typicode.com/posts"
response = requests.get(url)
api_data = pd.DataFrame(response.json())

print("\nDonnées récupérées depuis l'API :")
print(api_data.head())

# 6. Fusion des différentes sources de données dans un seul DataFrame
final_df = pd.concat([etudiants, excel_data, movies, api_data], axis=1)

print("\nDataFrame final après fusion des données :")
print(final_df.head())

# 7. Sauvegarde du DataFrame final dans un fichier CSV
final_df.to_csv("final_dataset.csv", index=False)

print("\nLe fichier 'final_dataset.csv' a été créé avec succès.")
```
